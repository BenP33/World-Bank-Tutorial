```python
#This file will contain all the code I use to get the data and organize it into a usable dataframe

#First we'll import the packages we'll need

import pandas as pd
import numpy as np

#this will allow us to call the API
import requests
import re
```

```python
#Structure the URL for the API call

#base for URL from the API documentation on the World Bank Website
url = "https://api.worldbank.org/v2/countries/all/indicator/DT.ODA.ODAT.CD?format=json"

r = requests.get(url, params={'date':'2011:2021',
                                'page':'1'})

r.status_code
```

```python
resultsDf = pd.DataFrame()

#the range will be 60, because we saw that there were 59 pages with results

for p in range(60):
    #iterate through each of the results pages
    req = requests.get(url, params={'date':'2011:2021','page':p})
    #add the data from the page to a temporary dataframe
    temp = pd.DataFrame(req.json()[1])
    #concate to the final dataframe
    resultsDf = pd.concat([resultsDf,temp])
```
